```python
import cv2
import numpy as np

#Define folder paths for where to find the images and where to save them
input_path = 'C:/Users/navid/OneDrive/Desktop/Enhancement/aligatorCracking01_Extracted/f_'
output_path = 'C:/Users/navid/OneDrive/Desktop/Enhancement/grayscale_aligator_Cracking01/gray_'

#Define the limit for how many frames to check
max_frames = 1000 
i = 0

#Start a loop to process each image one by one
while i < max_frames:

    #to get the full current filename
    current_file = input_path + str(i) + '.jpg'
    # Read the image from the computer
    image = cv2.imread(current_file)
    
    # Skip if the frame doesn't exist (deleted frames)
    if image is None:
        i = i + 1 
        continue  
    
    #convert to grayscale
    gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    #Create a new filename and save the grayscale image to the output folder
    save_file = output_path + str(i) + '.jpg'
    cv2.imwrite(save_file, gray_img)
    
    print("Frame " + str(i) + " converted to grayscale.")
    # Move to the next frame number
    i = i + 1

print("All found frames have been converted successfully!")
```
