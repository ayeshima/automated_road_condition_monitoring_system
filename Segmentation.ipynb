```python
import cv2
import numpy as np
import os

# 1. SETUP PATHS
input_folder = r'C:\Users\navid\OneDrive\Desktop\Enhancement\Final_enhancement\Final_enhanced_potHoles01'
output_root = r'C:\Users\navid\OneDrive\Desktop\Enhancement\Step_Outputs\Potholes_Only'

folders = {
    "otsu": os.path.join(output_root, "1_Otsu_Threshold"),
    "opening": os.path.join(output_root, "2_Morphological_Opening"),
    "contours": os.path.join(output_root, "3_Contour_Detection"),
    "final": os.path.join(output_root, "4_Final_Marked_Result")
}

for path in folders.values():
    os.makedirs(path, exist_ok=True)

valid_exts = ('.png', '.jpg', '.jpeg', '.bmp', '.tif')
count = 0

# 2. PROCESS IMAGES
images = [f for f in os.listdir(input_folder) if f.lower().endswith(valid_exts)]

for img_name in images:
    img_path = os.path.join(input_folder, img_name)
    img = cv2.imread(img_path, 0) 
    
    if img is None: continue

    # STEP 1: OTSU
    _, otsu_out = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    cv2.imwrite(os.path.join(folders["otsu"], f"otsu_{img_name}"), otsu_out)

    # STEP 2: OPENING
    kernel = np.ones((3, 3), np.uint8)
    opening_out = cv2.morphologyEx(otsu_out, cv2.MORPH_OPEN, kernel, iterations=1)
    cv2.imwrite(os.path.join(folders["opening"], f"open_{img_name}"), opening_out)

    # STEP 3: CONTOURS
    contours, _ = cv2.findContours(opening_out, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour_viz = np.zeros_like(img)
    cv2.drawContours(contour_viz, contours, -1, (255), 1)
    cv2.imwrite(os.path.join(folders["contours"], f"contour_{img_name}"), contour_viz)

    # STEP 4: FINAL
    final_result = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    cv2.drawContours(final_result, contours, -1, (0, 255, 0), 2)
    cv2.imwrite(os.path.join(folders["final"], f"final_{img_name}"), final_result)
    
    count += 1

print(f"Pothole processing done. Total: {count}")
```

```python
import cv2
import numpy as np
import os

# Set up the main input path
input_folder = r'C:\Users\navid\OneDrive\Desktop\Enhancement\Final_enhancement'

# Define where to save each step of the process
output_root = r'C:\Users\navid\OneDrive\Desktop\Enhancement\Step_Outputs'
folders = {
    "otsu": os.path.join(output_root, "1_Otsu_Threshold"),
    "opening": os.path.join(output_root, "2_Morphological_Opening"),
    "contours": os.path.join(output_root, "3_Contour_Detection"),
    "final": os.path.join(output_root, "4_Final_Marked_Result")
}

# Create all these folders if they aren't already there
for path in folders.values():
    if not os.path.exists(path):
        os.makedirs(path, exist_ok=True)

valid_exts = ('.png', '.jpg', '.jpeg', '.bmp', '.tif')
count = 0

# Loop through the category folders
for category in os.listdir(input_folder):
    cat_path = os.path.join(input_folder, category)
    
    if os.path.isdir(cat_path):
        print(f"Processing category: {category}")
        
        # Create sub-folders for the current category in each step folder
        for key in folders:
            os.makedirs(os.path.join(folders[key], category), exist_ok=True)

        # Get the images
        images = [f for f in os.listdir(cat_path) if f.lower().endswith(valid_exts)]

        for img_name in images:
            img_path = os.path.join(cat_path, img_name)
            img = cv2.imread(img_path, 0) # Load as grayscale
            
            if img is None:
                continue

            # --- STEP 1: OTSU THRESHOLDING ---
            # This turns the image black and white automatically
            _, otsu_out = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
            cv2.imwrite(os.path.join(folders["otsu"], category, f"otsu_{img_name}"), otsu_out)

            # --- STEP 2: MORPHOLOGICAL OPENING ---
            # Using a 3x3 kernel to remove tiny noise bits from the road
            kernel = np.ones((3, 3), np.uint8)
            opening_out = cv2.morphologyEx(otsu_out, cv2.MORPH_OPEN, kernel, iterations=1)
            cv2.imwrite(os.path.join(folders["opening"], category, f"open_{img_name}"), opening_out)

            # --- STEP 3: CONTOUR DETECTION ---
            # Finding the edges/lines of the damage
            contours, _ = cv2.findContours(opening_out, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            
            # Create a black background to show just the outlines for this step
            contour_viz = np.zeros_like(img)
            cv2.drawContours(contour_viz, contours, -1, (255), 1)
            cv2.imwrite(os.path.join(folders["contours"], category, f"contour_{img_name}"), contour_viz)

            # --- STEP 4: FINAL RESULT ---
            # Drawing the green outlines back onto the original image
            final_result = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
            cv2.drawContours(final_result, contours, -1, (0, 255, 0), 2)
            cv2.imwrite(os.path.join(folders["final"], category, f"final_{img_name}"), final_result)
            
            count += 1

print(f"\nDone! All steps saved in {output_root}")
print(f"Total images handled: {count}")
```
