```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Histogram Function
def plot_histogram(image, title):
    # Calculate how many pixels exist for each brightness level (0-255)
    histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
    plt.title(title)
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    # Draw the data as a blue line graph
    plt.plot(histogram, color='blue')
    plt.xlim([0, 256])
    
# Set where to find the grayscale images and where to save the analysis plots
input_path = 'C:/Users/navid/OneDrive/Desktop/Enhancement/grayscale_aligator_Cracking01/gray_'
output_path_plots = 'C:/Users/navid/OneDrive/Desktop/Enhancement/histogramPlots_aligatorCracking01/plot_'

max_frames = 1000
i = 0

# Keep running the code as long as the current number is less than 1000
while i < max_frames:

    # Create the filename and load the grayscale image
    current_file = input_path + str(i) + '.jpg'
    gray_img = cv2.imread(current_file, cv2.IMREAD_GRAYSCALE)

    # If the file doesn't exist, skip it and move to the next number
    if gray_img is None:
        i += 1
        continue
        
    # creat layout: left = images, right = histograms
    plt.figure(figsize=(12, 5))

    # Left Side → Grayscale Image
    plt.subplot(1, 2, 1)
    plt.imshow(gray_img, cmap='gray')
    plt.title('Grayscale Image (Frame ' + str(i) + ')')
    plt.axis('off')

    # Right Side → Histogram
    plt.subplot(1, 2, 2)
    plot_histogram(gray_img, "Original Histogram")

    # Adjust spacing so titles don't overlap
    plt.tight_layout()

    # Save plot
    plt.savefig(output_path_plots + str(i) + '.png')

    # Show on screen
    plt.show()

    # Close the window to save computer memory
    plt.close()

    print("Frame", i, "analyzed.")
    i += 1

print("All frames processed successfully!")
```
