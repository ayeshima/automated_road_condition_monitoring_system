```python
import cv2

#read the video
cap = cv2.VideoCapture('aligatorCracking01.mp4')
count = 0

#process the video while it is open
while cap.isOpened():
    #capture frame by frame
    ret,frame = cap.read()
    #check if the frame was read successfully if not stop the loop
    if not ret:break
        
    #save every 10th frame
    if count%10 == 0:
        # Save the current video frame as a JPEG image file with a unique number
        cv2.imwrite('C:/Users/navid/OneDrive/Desktop/Enhancement/f_%d.jpg' % count, frame)
        print('Saved:',count)
    #increment the frame counter
    count+=1
#close the video file
cap.release()
```
